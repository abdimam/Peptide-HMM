```python
import pandas as pd
import numpy as np
from tqdm import tqdm

def count_inversions_with_progressbar(reference_df, target_df):
    inversion_count = 0

    # Initialize tqdm with the total number of iterations
    total_iterations = len(target_df['sequence'])
    with tqdm(total=total_iterations, desc="Counting Inversions") as pbar:
        for seq_target in target_df['sequence']:
            # Create a boolean mask for sequences of the same length
            mask = reference_df['sequence'].str.len() == len(seq_target)

            # Extract sequences with the same length as seq_target
            sequences_same_length = reference_df.loc[mask, 'sequence'].values

            # Check lengths and continue if different
            if not sequences_same_length.any():
                continue

            # Iterate over sequences of the same length and check for switches
            for seq_reference in sequences_same_length:
                differences = np.array(list(seq_target)) != np.array(list(seq_reference))
                switches = np.logical_and(differences[:-1], differences[1:])
                
                # Count the number of inversions in each sequence
                inversion_count += np.sum(switches)

            # Update the progress bar
            pbar.update(1)

    return inversion_count

# Count all instances of inversions in mztab_data compared to used_set with progress bar
inversion_count = count_inversions_with_progressbar(used_set, mztab_data)

# Print the count of all instances of inversions
print(f"Total number of instances of inversions: {inversion_count}")
```
