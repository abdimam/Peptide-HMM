```python
import pandas as pd
from pyteomics import mztab
```

```python
df_targets = pd.read_parquet("targets/validation-00000-of-00001-b84568f5bf3ba95d.parquet")
df_validations = pd.read_parquet("targets/test-00000-of-00001-15be7bb3864037e7.parquet")

target = pd.concat([df_validations,df_targets], ignore_index=True)
target
```

```python
mztab_file_path = 'sequence_4_1_0_Saccharomyces_cerevisiae.mztab'
mztab_data = mztab.MzTab(mztab_file_path, table_format='df').spectrum_match_table
mztab_data['sequence'] = mztab_data['sequence'].apply(lambda seq: ''.join(char for char in seq if not char.isdigit() and char not in ['+', '-',"."]))
print(mztab_data.columns)
mztab_data
```

```python
# Extract 'ms_run' and 'index' using regular expressions
mztab_data[['ms_run', 'index']] = mztab_data['spectra_ref'].str.extract(r'ms_run\[(\d+)\]:index=(\d+)').astype(int)

# Create a new column with tuples (ms_run, index)
mztab_data['index_ms'] = list(zip(mztab_data['ms_run'], mztab_data['index']))

# Display the result
for index_ms in mztab_data["index_ms"]:
    print(index_ms)
```

```python
from tqdm import tqdm

length_msrun1 = list(range(27141+1))
length_msrun1_tuples = [(1, x) for x in length_msrun1]
length_msrun2 = list(range(28571+1))
length_msrun2_tuples = [(2, x) for x in length_msrun2]
length_msrun1_tuples = length_msrun1_tuples + length_msrun2_tuples
bool_list = []
#print(length_msrun1)
# Print the created list
#print(length_target)

for stuff in tqdm(length_msrun1_tuples, desc = "ms run and ID", unit = "row"):
    if (stuff) in tuple(mztab_data['index_ms']):
        bool_list.append(True)
        #print(stuff)
        #print("there")
    else:
        bool_list.append(False)
        print(stuff)
        print(" Not there")
        
        
    
    
```

```python
filtered_target = target.loc[bool_list]
filtered_target
```

```python
from collections import Counter

correct_counter = 0
correct_order = 0

for pred_seq, targ_seq in zip(mztab_data["sequence"], filtered_target["sequence"]):
    if Counter(pred_seq) == Counter(targ_seq):
        if pred_seq == targ_seq:
            correct_order += 1
        else:
            correct_counter += 1

ir = correct_counter / correct_order if correct_order > 0 else 0
print(ir)
            
```

```python
from tqdm import tqdm

pair_list = []
aa_score = {
    'A': 1,
    'R': 2,
    'N': 3,
    'D': 4,
    'C': 5,
    'Q': 6,
    'E': 7,
    'G': 8,
    'H': 9,
    'I': 10,  # Same score as L
    'L': 10,  # Same score as I
    'K': 11,
    'M': 12,
    'F': 13,
    'P': 14,
    'S': 15,
    'T': 16,
    'W': 17,
    'Y': 18,
    'V': 19,
}

def calculate_sequence_score(sequence):
    # Convert the sequence to uppercase to handle both upper and lower case letters
    sequence = sequence.upper()

    # Calculate the score for each amino acid and sum them up
    score = sum(aa_score.get(aa, 0) for aa in sequence)

    return score

correct = 0
not_correct = 0

# Assuming target and mztab_data are pandas DataFrames
target_sequences = target["sequence"].tolist()

# Create a tqdm progress bar for the outer loop
for sequence_target in tqdm(target_sequences, desc="Processing sequences", unit="sequence"):
    for sequence_pred in mztab_data["sequence"]:
        if len(sequence_target) == len(sequence_pred):
            if calculate_sequence_score(sequence_target) == calculate_sequence_score(sequence_pred):
                if sequence_target == sequence_pred:
                    correct += 1
                else:
                    not_correct += 1

print("Correct:", correct)
print("Not Correct:", not_correct)
                
            
        
```
